```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# Load the dataset (assume preprocessed)
data = pd.read_csv('/Users/mage/Documents/graduates study/visualization/Chicago crime/chicago_crime_data.csv')
data['Date'] = pd.to_datetime(data['Date'])

# Filter data for 2023 and focus on 4 crime types
filtered_data = data[(data['Date'] >= '2023-01-01') & (data['Date'] <= '2023-06-30')]
crime_types_of_interest = ['BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'THEFT']
filtered_data = filtered_data[filtered_data['Primary Type'].isin(crime_types_of_interest)]

# Prepare data for bar plot (Crimes by District)
crime_by_district = filtered_data.groupby('District').size().sort_values(ascending=False)

# Prepare data for small panels
# 1. Crimes by Month
filtered_data['Month'] = filtered_data['Date'].dt.month
crime_by_month = filtered_data.groupby('Month').size()

# 2. Arrests by Crime Type
arrests_by_crime = filtered_data.groupby('Primary Type')['Arrest'].sum()

# 3. Domestic incidents by Crime Type
domestic_by_crime = filtered_data.groupby('Primary Type')['Domestic'].sum()

# Create the composite figure
fig = plt.figure(figsize=(16, 12))
gs = GridSpec(3, 2, figure=fig, width_ratios=[3, 1], height_ratios=[2, 1, 1])

# Main plot: Scatter plot of crime locations
ax1 = fig.add_subplot(gs[0:2, 0])  # Main panel spans two rows
sns.scatterplot(
    x='Longitude', 
    y='Latitude', 
    hue='Primary Type', 
    data=filtered_data.sample(frac=0.1, random_state=42),  # Sample for speed
    alpha=0.6,
    s=10,
    ax=ax1
)
ax1.set_title('Crime Locations in Chicago (Jan-Jun 2023)')
ax1.set_xlabel('Longitude')
ax1.set_ylabel('Latitude')
ax1.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Side panel 1: Bar plot of crimes by district
ax2 = fig.add_subplot(gs[0, 1])
sns.barplot(
    x=crime_by_district.index.astype(str), 
    y=crime_by_district.values, 
    palette='coolwarm',
    ax=ax2
)
ax2.set_title('Crimes by District')
ax2.set_xlabel('District')
ax2.set_ylabel('Number of Crimes')
ax2.tick_params(axis='x', rotation=45)

# Side panel 2: Crimes by month
ax3 = fig.add_subplot(gs[1, 1])
sns.barplot(
    x=crime_by_month.index.astype(str), 
    y=crime_by_month.values, 
    palette='viridis',
    ax=ax3
)
ax3.set_title('Crimes by Month')
ax3.set_xlabel('Month')
ax3.set_ylabel('Number of Crimes')

# Side panel 3: Arrests by crime type
ax4 = fig.add_subplot(gs[2, 0])
sns.barplot(
    x=arrests_by_crime.index, 
    y=arrests_by_crime.values, 
    palette='mako',
    ax=ax4
)
ax4.set_title('Arrests by Crime Type')
ax4.set_xlabel('Crime Type')
ax4.set_ylabel('Number of Arrests')
ax4.tick_params(axis='x', rotation=45)

# Side panel 4: Domestic incidents by crime type
ax5 = fig.add_subplot(gs[2, 1])
sns.barplot(
    x=domestic_by_crime.index, 
    y=domestic_by_crime.values, 
    palette='rocket',
    ax=ax5
)
ax5.set_title('Domestic Incidents by Crime Type')
ax5.set_xlabel('Crime Type')
ax5.set_ylabel('Number of Incidents')
ax5.tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig('composite_figure.png', dpi=300, bbox_inches='tight')  # Save with high resolution
plt.show()
```
